## 3. ***Data Wrangling***
### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
# 1. Convert Timestamp to datetime format
df['Timestamp'] = pd.to_datetime(df['Timestamp'])
````

````python
# 2. Age column contains extreme and negative values like -29, 329, 99999999999, -1726, 5, 8, 11.
df = df[(df['Age'] >= 16) & (df['Age'] <= 100)]
````

````python
# 3. Gender column contains 49 unique values due to typos, casing, and gender diversity.  Normalize to main categories: 'Male', 'Female', 'Other'.
def clean_gender(gender):
    gender = str(gender).strip().lower()
    if gender in ['male', 'm', 'man', 'male ', 'cis male', 'cis man', 'mal', 'msle', 'mail', 'malr', 'maile']:
        return 'Male'
    elif gender in ['female', 'f', 'woman', 'female ', 'cis female', 'femake', 'femail']:
        return 'Female'
    else:
        return 'Other'

df.loc[:, 'Gender'] = df['Gender'].apply(clean_gender)
````

````python
# 4. columns self_employed and work_interference contains missing values. Fill missing values
df = df[(df['Age'] >= 16) & (df['Age'] <= 100)].copy()  # ✅ makes a new DataFrame safely

df['self_employed'] = df['self_employed'].fillna('Not specified')
df['work_interfere'] = df['work_interfere'].fillna('Unknown')
````

````python
# 5. no_employees columns - group into categories like small, medium and large
size_map = {
    '1-5': 'Small', '6-25': 'Small',
    '26-100': 'Medium', '100-500': 'Medium',
    '500-1000': 'Large', 'More than 1000': 'Large'
}
df['no_employees_group'] = df['no_employees'].map(size_map)
````

````python
# 6. column leave - responses are grouped for simplicity
df['leave_grouped'] = df['leave'].replace({
    'Very easy': 'Easy',
    'Somewhat easy': 'Easy',
    'Somewhat difficult': 'Difficult',
    'Very difficult': 'Difficult',
    "Don't know": "Uncertain"
})
````

### What all manipulations have you done and insights you found?
As part of data wrangling and exploration, I performed several preprocessing steps on selected columns to ensure consistency, remove invalid entries, and prepare the dataset for analysis. Below are the specific manipulations and associated insights:

1. Timestamp
Manipulation: Converted the Timestamp column to datetime format using pd.to_datetime() to enable temporal analysis.

Insight: The dataset includes survey responses spanning from 2014 to 2016, which could be useful to track any trends over time in attitudes toward mental health.

2. Age
Manipulation: Identified and removed outliers and unrealistic age values such as -1726, 329, and 99999999999. Retained only values in the range 16 to 100, which are more representative of the working population.

Insight: The majority of respondents fall between the ages of 25 and 40, suggesting that the data primarily represents early-to-mid-career professionals in the tech sector.

3. Gender
Manipulation: Cleaned and standardized gender values by grouping over 40 variants into three broad categories: Male, Female, and Other. This involved fixing inconsistent casing, spelling errors, and inclusive identity labels.

Insight: While most respondents identified as male or female, a small but significant portion selected non-binary or diverse gender identities. This highlights the need to ensure inclusivity in workplace mental health policies.

4. Self-employed
Manipulation: Filled missing values with "Not specified" to retain completeness and avoid discarding potentially valuable data.

Insight: Most respondents are not self-employed. Those who are may face different challenges in accessing mental health resources due to lack of formal organizational support.

5. Work Interfere
Manipulation: Filled missing values with "Unknown" assuming that either the question didn’t apply or was skipped. This preserved the categorical integrity of the column.

Insight: A considerable number of respondents report that mental health issues “Sometimes” or “Often” interfere with their work, suggesting the need for better workplace support and accommodations.

6. No_employees
Manipulation: Grouped company sizes into three broader categories: Small (1–25), Medium (26–500), and Large (500+), to simplify comparison across company scales.

Insight: Most participants are from medium to large companies. Larger organizations may offer more structured mental health programs, but attitudes still vary significantly.

7. Leave
Manipulation: Simplified the values into categories such as Easy, Difficult, and Uncertain by grouping similar responses together.

Insight: A large number of respondents find it difficult or somewhat difficult to take medical leave for mental health, highlighting stigma or procedural barriers in the workplace.

#### Chart - 14 - Correlation Heatmap

````python
# Correlation Heatmap visualization code
# Make a copy and encode only valid binary columns
df_encoded = df.copy()

# Map Yes/No to 1/0
binary_cols = ['treatment', 'family_history', 'tech_company', 'benefits',
               'care_options', 'wellness_program', 'seek_help', 'anonymity',
               'mental_health_consequence', 'phys_health_consequence', 'obs_consequence']

for col in binary_cols:
    df_encoded[col] = df_encoded[col].map({'Yes': 1, 'No': 0})

# Correlation matrix
corr = df_encoded.corr(numeric_only=True)

# Plot heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap of Mental Health Factors")
plt.savefig("C:\\Users\\Raghu\\OneDrive\\Desktop\\Mental_Health_Data\\correlation_heatmap.png", dpi=300)
plt.show()
````

##### 1. Why did you pick the specific chart?
A correlation heatmap, which shows relationships (linear correlation) between numeric variables.

##### 2. What is/are the insight(s) found from the chart?
 1. seek_help is strongly correlated with wellness_program (0.72): Employees are much more likely to seek help for mental health if their employer has a wellness program.
 2. seek_help also correlates with benefits (0.55) and care_options (0.39): When mental health benefits and care options are available, employees feel more empowered to seek help.
 3. anonymity is negatively correlated with mental_health_consequence (-0.49): Where employees feel their anonymity is protected, they are less concerned about negative consequences of discussing mental health.

#### Chart - 15 - Pair Plot

````python
# Pair Plot visualization code
# First, encode relevant categorical variables for numeric comparison
df_pair = df.copy()
df_pair['treatment'] = df_pair['treatment'].map({'Yes': 1, 'No': 0})
df_pair['family_history'] = df_pair['family_history'].map({'Yes': 1, 'No': 0})
df_pair['tech_company'] = df_pair['tech_company'].map({'Yes': 1, 'No': 0})
df_pair['Gender'] = df_pair['Gender'].apply(lambda x: 'Male' if 'male' in x.lower() else 'Female' if 'female' in x.lower() else 'Other')

# Now plot the pairplot
sns.pairplot(df_pair, vars=['Age', 'treatment', 'family_history'],
             hue='Gender', palette='Set2', diag_kind='kde')
plt.suptitle('Pairplot of Age, Family History & Treatment by Gender', y=1.02)
plt.show()
````

Pairplots are excellent for simultaneously showing:Distributions of individual variables (diag_kind='kde'), Relationships between multiple pairs and Groupings by color (hue='Gender')
